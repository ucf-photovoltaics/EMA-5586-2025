```r
files <- fs::dir_ls("Data_Complex_Refractive_Index/", glob = "*.csv")

complex_refractive_index <- read_csv(files, id = "filename")

# complex_refractive_index <- list.files(path = "complex_refractive_index_data/", pattern = "*.csv", full.names = TRUE) %>%
#	lapply(read_csv) %>%
#	bind_rows

complex_refractive_index_df <- data.frame(complex_refractive_index)

colnames(complex_refractive_index_df) [1] <- "material_ID"
colnames(complex_refractive_index_df) [2] <- "wavelength_n"
colnames(complex_refractive_index_df) [4] <- "wavelength_k"
colnames(complex_refractive_index_df) [6] <- "alpha"

complex_refractive_index_df

str(complex_refractive_index_df)

options(repr.plot.width = 10, repr.plot.height = 8)
```

```r
# Plot of the real refractive index versus wavelength
ggplot(data = complex_refractive_index_df, aes(y = n, x = wavelength_n, group = material_ID, color = material_ID)) +
  	geom_line() +
	theme_bw() +
  	theme(text = element_text(size = 30)) +
  	theme(axis.ticks.length = unit(-0.25, "cm")) +
	scale_color_discrete(name = "Material", labels = c("A", "B", "C", "D", "E")) +
	ylab(expression(italic(n))) + 
	xlab(expression(italic(λ)~(nm))) +
	xlim(300, 1300)

# Plot of the imaginary refractive index versus wavelength
ggplot(data = complex_refractive_index_df, aes(y = k, x = wavelength_k, group = material_ID, color = material_ID)) +
  	geom_line() +
	theme_bw() +
  	theme(text = element_text(size = 30)) +
  	theme(axis.ticks.length = unit(-0.25, "cm")) +
	scale_color_discrete(name = "Material", labels = c("A", "B", "C", "D", "E")) +
	ylab(expression(italic(k))) + 
	xlab(expression(italic(λ)~(nm))) +
	ylim(0.01, 1.1) +
	xlim(300, 1300) +
	scale_y_log10(breaks = trans_breaks("log10", function(x) 10^x), labels = trans_format("log10", math_format(10^.x)), limits = c(0.01, 5)) 
```
